# Leave One-Out cross validation (LOO)

```python
in LOO cross validation, dataset is divided in such a way that if there are total n rows in our dataset
then it is divided "n" times and perform n iterations , in such a way that if n=10 then it is trained on 9 rows
and tested on remaining 1 and in this way it perform total 10 iterations in which in 10th iteration testing or validation
will occur on 10th row.

    but there is one problem with it is that if the dataset is big, it takes a lot of time, so this is good only for
small datasets.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('placement.csv', nrows=100)
df.head() 
```

```python
x=df[['cgpa']]
y=df['package']
```

```python
sns.scatterplot(x='cgpa', y='package', data=df)
plt.show()
```

## to check how data is splitted with different cross validation methods

```python
from sklearn.model_selection import KFold, LeaveOneOut, LeavePOut, StratifiedKFold
kf=LeaveOneOut()
kf=LeavePOut(p=2)
kf=KFold(n_splits=6)

for train, test in kf.split(x,y):
    print('training parts are',train)
    print('testing parts are',test)
    print('\n')
```

## how to check score using cross validation functions

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
score=cross_val_score(LinearRegression(),x,y, cv=5) # on direclty passing cv=5 , by default 5 fold cross validation is used, but
# if you want to select metod on its own, you dio it like below:
score=cross_val_score(LinearRegression(), x,y, cv=KFold(n_splits=5))
score.sort()
score
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
lr=LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
It's normal to see small variations in accuracy, and an increase from 82% to 85% could 
simply be due to a more favorable split of the data in the second method. 
```
